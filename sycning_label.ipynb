```python
import pandas as pd
import numpy as np

def interpolate_labels(data_file, label_file, output_file):
    # Đọc file dữ liệu (không có header, có 7 cột)
    data = pd.read_csv(data_file, sep=',', header=None, names=['timestamp', 'x1', 'x2', 'x3', 'x4', 'x5', 'id_label'])

    # Đọc file label (có header, 2 cột)
    labels = pd.read_csv(label_file)

    # Nội suy giá trị label dựa trên timestamp
    labels['label'] = labels['label'].astype(float)
    interpolated_labels = np.interp(data['timestamp'], labels['timestamp'], labels['label'])

    # Gán lại giá trị id_label cho data
    data['id_label'] = interpolated_labels

    # Lưu dữ liệu mới vào file output
    data.to_csv(output_file, sep=',', index=False, header=False)

# Đường dẫn tới các file
input_data_file = "1734510152055-1734510164061.csv"
input_label_file = "1734510152055-1734510164061_label.csv"
output_file = "output.csv"

# Gọi hàm xử lý
interpolate_labels(input_data_file, input_label_file, output_file)
```

```python
import pandas as pd

def nearest_label_interpolation(data_file, label_file, output_file):
    # Đọc file dữ liệu (không có header, có 7 cột)
    data = pd.read_csv(data_file, sep=',', header=None, names=['timestamp', 'x1', 'x2', 'x3', 'x4', 'x5', 'id_label'])
    
    # Đọc file label (có header, 2 cột)
    labels = pd.read_csv(label_file)
    
    # Đảm bảo cột timestamp có kiểu dữ liệu đồng nhất (float)
    data['timestamp'] = data['timestamp'].astype(float)
    labels['timestamp'] = labels['timestamp'].astype(float)
    
    # Tìm nhãn gần nhất ở phía trước hoặc trùng khớp
    def find_nearest_label(timestamp):
        nearest_label_row = labels[labels['timestamp'] <= timestamp]
        if nearest_label_row.empty:
            return None  # Không tìm thấy nhãn phù hợp
        return nearest_label_row.iloc[-1]['label']
    
    # Áp dụng hàm để tính nhãn mới
    data['id_label'] = data['timestamp'].apply(find_nearest_label)
    
    # Lưu dữ liệu mới vào file output
    data.to_csv(output_file, sep=',', index=False, header=False)

# Đường dẫn tới các file
input_data_file = "1734510152055-1734510164061.csv"
input_label_file = "1734510152055-1734510164061_label.csv"
output_file = "output.csv"

# Gọi hàm xử lý
nearest_label_interpolation(input_data_file, input_label_file, output_file)
```

```python
import pandas as pd
import os

def nearest_label_interpolation(data_file, label_file, output_file):
    # Đọc file dữ liệu (không có header, có 7 cột)
    data = pd.read_csv(data_file, sep=',', header=None, names=['timestamp', 'x1', 'x2', 'x3', 'x4', 'x5', 'id_label'])
    
    # Đọc file label (có header, 2 cột)
    labels = pd.read_csv(label_file)
    
    # Đảm bảo cột timestamp có kiểu dữ liệu đồng nhất (float)
    data['timestamp'] = data['timestamp'].astype(float)
    labels['timestamp'] = labels['timestamp'].astype(float)
    
    # Tìm nhãn gần nhất ở phía trước hoặc trùng khớp
    def find_nearest_label(timestamp):
        nearest_label_row = labels[labels['timestamp'] <= timestamp]
        if nearest_label_row.empty:
            return None  # Không tìm thấy nhãn phù hợp
        return f"id{int(nearest_label_row.iloc[-1]['label'])}"
    
    # Áp dụng hàm để tính nhãn mới
    data['id_label'] = data['timestamp'].apply(find_nearest_label)
    
    # Lưu dữ liệu mới vào file output
    data.to_csv(output_file, sep=',', index=False, header=False)

def process_folder(folder_path):
    # Lấy danh sách các file trong thư mục
    files = [f for f in os.listdir(folder_path) if f.endswith('.csv') and not f.endswith('_label.csv')]
    
    for file in files:
        base_name = file.split('.')[0]
        data_file = os.path.join(folder_path, file)
        label_file = os.path.join(folder_path, f"{base_name}_label.csv")
        output_file = os.path.join(folder_path, f"{base_name}_synced.csv")
        
        if os.path.exists(label_file):
            print(f"Processing: {file} with {label_file}")
            nearest_label_interpolation(data_file, label_file, output_file)
        else:
            print(f"Label file not found for {file}")

# Đường dẫn tới thư mục chứa dữ liệu
folder_path = r"."
process_folder(folder_path)
```
